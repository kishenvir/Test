```python
def isPalindrome(s): 
      
    # Using predefined function to  
    # reverse to string print(s) 
    rev = ''.join(reversed(s)) #reversed() built in function used with join() built in functon to reverse and handle the value as string('')
    print(rev)
  
    # Checking if both string are  
    # equal or not 
    if (s == rev): 
        return True
    return False
  
# main function 
s = "one"
ans = isPalindrome(s) 
  
if (ans): 
    print("Yes") 
else: 
    print("No") 
```

```python
#Write a Python function that takes a number as a parameter and #check the number is prime or not

def test_prime(n):
    if (n==1):
        return False
    elif (n==2):
        return True;
    else:
        for x in range(2,n):
            if(n % x==0):
                return False
        return True             
print(test_prime(18))
```
